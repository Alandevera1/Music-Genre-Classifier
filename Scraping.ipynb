```python
# Importing libraries...
import pandas as pd
import requests
import bs4
```

```python
# Requesting data from everynoise.com  
URL = 'https://everynoise.com/everynoise1d.cgi?scope=all'
response = requests.get(URL)
response
```

```python
# Parsing through html text using BeautifulSoup
soup = bs4.BeautifulSoup(response.text, "lxml")
soup
```

```python
# Sifting through html to get links to the albums 
hrefs = soup.find_all('a', {'class', 'note'})
hrefs[0].get('href')
```

```python
# Sifting through html to get the names of the genres
genres = soup.find_all({'td', 'class'}, {'note','title'})
genres = genres[1::2] # slicing every odd index
genres[0].text
```

```python
# Put data into a DataFrame and csv
df = pd.DataFrame(columns=['genre','href'])
df.genre = [genre.text for genre in genres]
df.href = [href.get('href') for href in hrefs]

df.to_csv('SpotifyGenrePlaylists.csv')
```
