```python
# Importing libraries...

import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
genres = pd.read_csv('genres.csv')
genres = genres.dropna()
genres.head()
```

```python
# top_genres = genres.loc[genres.genre.isin(['pop', 'rock', 'rap', 'hip hop', 'edm', 'r&b', 'jazz', 'country', ' metal', 'indie pop', 'classical'])]
# top_genres = genres.loc[genres.genre.isin(genres.genre.unique()[0:50])]
top_genres = genres.loc[genres.genre.isin(['pop', 'rock', 'rap', 'hip hop', 'edm', 'r&b', 'jazz', 'country', ' metal', 'indie rock', 'classical'])]

top_genres.genre.unique()
```

```python
x = pd.DataFrame(top_genres[['acousticness','danceability','energy','instrumentalness','liveness','loudness','speechiness','tempo','valence']])
y = pd.DataFrame(top_genres.genre, columns=['genre'])
y
```

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split

(x_train, x_test, y_train, y_test) = train_test_split(x, y, test_size=0.2)
y_train.shape
```

```python
tree = DecisionTreeClassifier(criterion="entropy")
tree.fit(x.values, np.array(y).reshape(len(y),))

y_pred = tree.predict(x_test.values)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
sample_one = tree.predict([[0.235, 0.572, 0.743, 0.00000593, 0.0761, -3.991, 0.12, 170.123, 0.678]])
sample_two = tree.predict([[0.0139, 0.703, 0.834, 0, 0.1, -5.134, 0.0603, 153.97, 0.863]])
sample_three = tree.predict([[0.876, 0.517, 0.0917, 0.629, 0.108, -18.395, 0.0336, 95.068, 0.0836]])
sample_four = tree.predict([[0.00764, 0.421, 0.958, 0.269, 0.103, -2.886, 0.0755, 118.192, 0.496]])

print("[Red Velvet - Kingdom Come] is most likely of genre", sample_one)
print("[Denzel Curry - Dynasties & Dystopia] is most likely of genre", sample_two)
print("[Esbjorn Svensson Trio - Where We Used to Live] is most likely of genre", sample_three)
print("[Thank You Scientist - Life of Vermin] is most likely of genre", sample_four)
```

```python
from sklearn.svm import SVC

svclassifier = SVC(kernel="linear")
svclassifier.fit(x_train.values, np.array(y_train).reshape(len(y_train),))
y_pred = svclassifier.predict(x_test.values)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(criterion='entropy', n_estimators=100, min_samples_split=11, max_features='auto')
rfc.fit(x_train.values, np.array(y_train).reshape(len(y_train),))
y_pred = rfc.predict(x_test.values)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
